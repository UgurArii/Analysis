```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 11
x = np.linspace(0, 10, N)
print(x)


y = x
plt.plot(x, y, 'o--')
plt.axis('off')
plt.show()
```

```python
x = np.arange(5)
y = x
plt.plot(x, y+1, 'g')
plt.plot(x, y+0.5, 'y')
plt.plot(x, y, 'r')
plt.plot(x, y-0.2, 'c')
plt.plot(x, y-0.4, 'k')
plt.plot(x, y-0.6, 'm')
plt.plot(x, y-0.8, 'w')
plt.plot(x, y-1, 'b')
plt.show()
```

```python
plt.plot(x, y+1, 'g', x, y+0.5, 'y', x, y, 'r', x, y-0.2, 'c', 
x, y-0.4, 'k', x, y-0.6, 'm', x, y-0.8, 'w', x, y-1, 'b')
plt.show()
```
